```python
def mycustom_boundary(docx):
    for token in docx[:-1]:
        if token.text == '.':
            docx[token.i+1].is_sent_start=True
    return docx
```

```python
nlp.add_pipe(mycustom_boundary,before='parser')
for num,sentence in enumerate(doc_file.sents):
    print(f'{num}:{sentence}')
```
